```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('penguins')

# 날개의 길이와 몸무게
plt.scatter(df['flipper_length_mm'], df['body_mass_g'])
plt.show()
```

```python
df_group = df.groupby('species')['body_mass_g'].mean().reset_index()
df_group
```

```python
plt.bar(x=df_group['species'], height=df_group['body_mass_g'])
plt.show()
```

```python
plt.hist(df['body_mass_g'], bins=30)
plt.xlabel('Body Mass')
plt.ylabel('Count')
plt.title('distribution')
plt.show()
```

```python
import pandas as pd

# 미국 실업자 데이터
df_unrate = pd.read_csv(
    'https://research.stlouisfed.org/fred2/series/UNRATE/downloaddata/UNRATE.csv'
)
df_unrate.head()
```

```python
df_unrate['DATE'] = pd.to_datetime(df_unrate['DATE'])

plt.plot(df_unrate['DATE'], df_unrate['VALUE'])
plt.show()
```

```python


# 행이 2개, 열이 1개
# 가로 세로 길이는 10, 6
fig, axes = plt.subplots(2, 1, figsize=(10, 6))

# 첫번째 그림
axes[0].scatter(df['flipper_length_mm'], df['body_mass_g'])
axes[0].set_xlabel('mm')
axes[0].set_ylabel('g')
axes[0].set_title('relation')

# 두번째 그림
axes[1].hist(df['body_mass_g'], bins=30)
axes[1].set_xlabel('body mass')
axes[1].set_ylabel('count')
axes[1].set_title('distribution')

# 간격 조정
plt.subplots_adjust(
    left=0.1,
    right=0.95,
    bottom=0.1,
    top=0.95,
    wspace=0.5,
    hspace=0.5
)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# 첫번째 그림
plt.subplot(2, 1, 1)
plt.scatter(df['flipper_length_mm'], df['body_mass_g'])
plt.xlabel('mm')
plt.ylabel('g')
plt.title('relation')

# 두번째 그림
plt.subplot(2, 1, 2)
plt.hist(df['body_mass_g'], bins=30)
plt.xlabel('body mass')
plt.ylabel('count')
plt.title('distribution')

# 간격 조정
plt.subplots_adjust(
    left=0.1,
    right=0.95,
    bottom=0.1,
    top=0.95,
    wspace=0.5,
    hspace=0.5
)
plt.show()
```
